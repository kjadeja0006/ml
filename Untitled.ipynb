```python
# Q-1
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("1.jpeg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img2 = cv2.Canny(img, 50 , 150 )

plt.subplot(1,1,1)
plt.imshow(img2, cmap="gray")
plt.axis('off')
```

```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
# nltk.download()

read = open("2.txt").read()
output=nltk.word_tokenize(read)
print(output)

from nltk.stem.porter import PorterStemmer

words=output
stem=PorterStemmer()

for word in words:
    stem_words=[stem.stem(word)]
    
print("Porter Stemmer ==== ",word,stem_words)    
```
